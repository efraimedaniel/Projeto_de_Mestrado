```python
import matplotlib.pyplot as plt
import pandas as pd

# Extrair apenas os anos do seu conjunto de dados
sz_imbituba['ano'] = pd.to_datetime(sz_imbituba['data_hora']).dt.year
anos = sorted(set(sz_imbituba['ano']))

plt.figure(figsize=(12, 6))
plt.plot(sz_imbituba['anomes'], sz_imbituba['Media_Movel_Centrada']*100)
plt.grid(True)
plt.xlabel('Anos')
plt.ylabel('Amplitude (cm)', fontweight='bold')
plt.title('Variabilidade interanual - Imbituba', fontsize=16, fontweight='bold')

# Exibir rótulos de ano a cada intervalo de 5 anos
anos_exibir = anos[::2]
indices_exibir = [anos.index(ano) * 12 for ano in anos_exibir]
plt.xticks(indices_exibir, anos_exibir)
# Exibir apenas os anos no eixo x
#plt.xticks(range(0, len(anos)*12, 12), anos, rotation=45)  # Corrigido aqui

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Tamanho da folha A4 em polegadas
largura_a4_inches = 8.27  # largura em polegadas
altura_a4_inches = 11.69  # altura em polegadas

# Resolução desejada
dpi = 300

# Criar a figura e os subplots
fig, axs = plt.subplots(7, 1, figsize=(largura_a4_inches, altura_a4_inches), dpi=dpi)

# Imbituba
sz_imbituba['ano'] = pd.to_datetime(sz_imbituba['data_hora']).dt.year
anos_imbituba = sorted(set(sz_imbituba['ano']))
axs[0].plot(sz_imbituba['anomes'], sz_imbituba['Media_Movel_Centrada']*100, 'r')
#axs[0].grid(True)
axs[0].set_xlabel('Anos')
axs[0].set_ylabel('Amplitude (cm)', fontsize=8)
#axs[0].set_title('Variabilidade Interanual', fontsize=12, fontweight='bold')
axs[0].legend(['Imbituba'], loc='upper left')
anos_exibir = anos_imbituba[::2]
indices_exibir = [anos_imbituba.index(ano) * 12 for ano in anos_exibir]
axs[0].set_xticks(indices_exibir)
axs[0].set_xticklabels(anos_exibir)
axs[0].set_ylim(-10 , 10)
axs[0].set_yticks(range(-10, 11, 5))




# Cananéia
sz_cananeia['ano'] = pd.to_datetime(sz_cananeia['data_hora']).dt.year
anos_cananeia = sorted(set(sz_cananeia['ano']))
axs[1].plot(sz_cananeia['anomes'], sz_cananeia['Media_Movel_Centrada']*100, 'g')
#axs[1].grid(True)
axs[1].set_xlabel('Anos')
axs[1].set_ylabel('Amplitude (cm)', fontsize=8)
#axs[1].set_title('Variabilidade interanual - Cananéia', fontsize=16, fontweight='bold')
axs[1].legend(['Cananeia'], loc='upper left')
anos_exibir = anos_cananeia[::10]
indices_exibir = [anos_cananeia.index(ano) * 12 for ano in anos_exibir]
axs[1].set_xticks(indices_exibir)
axs[1].set_xticklabels(anos_exibir)
#margin = 4  # Margem para incluir espaço entre os valores e as bordas do gráfico
axs[1].set_ylim(-20 , 20)
axs[1].set_yticks(range(-20, 21, 10))

# Arraial
sz_arraial['ano'] = pd.to_datetime(sz_arraial['dt_hr']).dt.year
anos_arraial = sorted(set(sz_arraial['ano']))
axs[2].plot(sz_arraial['anomes'], sz_arraial['Media_Movel_Centrada']*100, 'b')
#axs[2].grid(True)
axs[2].set_xlabel('Anos')
axs[2].set_ylabel('Amplitude (cm)', fontsize=8)
#axs[2].set_title('Variabilidade interanual - Arraial', fontsize=16, fontweight='bold')
axs[2].legend(['Arraial do cabo'], loc='upper left')
anos_exibir = anos_arraial[::]
indices_exibir = [anos_arraial.index(ano) * 12 for ano in anos_exibir]
axs[2].set_xticks(indices_exibir)
axs[2].set_xticklabels(anos_exibir)
#margin = 1
axs[2].set_ylim(26, 30)
axs[2].set_yticks(range(26, 31, 1))


# Salvador
sz_salvador['ano'] = pd.to_datetime(sz_salvador['data_hora']).dt.year
anos_salvador = sorted(set(sz_salvador['ano']))
axs[3].plot(sz_salvador['anomes'], sz_salvador['Media_Movel_Centrada']*100, 'c')
#axs[3].grid(True)
axs[3].set_xlabel('Anos')
axs[3].set_ylabel('Amplitude (cm)', fontsize=8)
#axs[3].set_title('Variabilidade interanual - Salvador', fontsize=16, fontweight='bold')
axs[3].legend(['Salvador'], loc='upper left')
anos_exibir = anos_salvador[::2]
indices_exibir = [anos_salvador.index(ano) * 12 for ano in anos_exibir]
axs[3].set_xticks(indices_exibir)
axs[3].set_xticklabels(anos_exibir)
axs[3].set_ylim(36, 46) 
axs[3].set_yticks(range(37, 47, 2))

#Fortaleza
sz_fortaleza['ano'] = pd.to_datetime(sz_fortaleza['data_hora']).dt.year
anos_salvador = sorted(set(sz_salvador['ano']))
axs[4].plot(sz_fortaleza['anomes'], sz_fortaleza['Media_Movel_Centrada']*100, 'm')
#axs[4].grid(True)
axs[4].set_xlabel('Anos')
axs[4].set_ylabel('Amplitude (cm)', fontsize=8)
#axs[3].set_title('Variabilidade interanual - Salvador', fontsize=16, fontweight='bold')
axs[4].legend(['Fortaleza'], loc='best')
anos_exibir = anos_salvador[::2]
indices_exibir = [anos_salvador.index(ano) * 12 for ano in anos_exibir]
axs[4].set_xticks(indices_exibir)
axs[4].set_xticklabels(anos_exibir)
#axs[4].set_ylim(46, 54)
axs[4].set_ylim(46, 54) 
axs[4].set_yticks(range(46, 55, 2))

# Belém
sz_belem['ano'] = pd.to_datetime(sz_belem['time_02']).dt.year
anos_belem = sorted(set(sz_belem['ano']))
axs[5].plot(sz_belem['anomes'], sz_belem['Media_Movel_Centrada']*100, 'y')
#axs[5].grid(True)
axs[5].set_xlabel('Anos')
axs[5].set_ylabel('Amplitude (cm)', fontsize=8)
#axs[4].set_title('Variabilidade interanual - Belém', fontsize=16, fontweight='bold')
axs[5].legend(['Belém'], loc='upper left')
anos_exibir = anos_belem[::]
indices_exibir = [anos_belem.index(ano) * 12 for ano in anos_exibir]
axs[5].set_xticks(indices_exibir)
axs[5].set_xticklabels(anos_exibir)
axs[5].set_ylim(110, 116) 
axs[5].set_yticks(range(110, 117, 2))

# Santana
sz_santana['ano'] = pd.to_datetime(sz_santana['dt_hr']).dt.year
anos_santana = sorted(set(sz_santana['ano']))
axs[6].plot(sz_santana['Media_Movel_Centrada']*100, 'k')
#axs[6].grid(True)
axs[6].set_xlabel('Anos')
axs[6].set_ylabel('Amplitude (cm)', fontsize=8)
#axs[6].set_title('Variabilidade interanual - Santana', fontsize=16, fontweight='bold')
axs[6].legend(['Santana'], loc='upper left')
anos_exibir = anos_santana[::2]
indices_exibir = [anos_santana.index(ano) * 12 for ano in anos_exibir]
axs[6].set_xticks(indices_exibir)
axs[6].set_xticklabels(anos_exibir)
axs[6].set_ylim(15, 55) 
axs[6].set_yticks(range(15, 56, 10))

# Ajuste de layout
#plt.tight_layout()

# Ajuste de layout
plt.tight_layout()

# Salvar a figura na pasta especificada
plt.savefig('D:/Sazonalidade/dataframe_sazonalidades/interanual_maré_semgrid.png')  # Salva como 'interanual_maré_02.png'

# Mostrar a figura
plt.show()
```
